```python
#Get pandas data
from google.colab import files
import pandas as pd

# Upload the file
uploaded = files.upload()

# 'uploaded' is a dictionary containing the names and contents of the uploaded files
# Get the name of the uploaded file
file_name = list(uploaded.keys())[0]

# Read the CSV file into a DataFrame
fandango_df = pd.read_csv(file_name)

# Display the DataFrame
print(fandango_df)

```

```python
corr_coefficient = fandango_df.drop(columns='FILM').corr()['VOTES']['RATING']

plt.figure(figsize=(12, 5))
sns.scatterplot(data=fandango_df, x=fandango_df['RATING'], y=fandango_df['VOTES'])

plt.text(0.05, 0.95, f'Correlation: {corr_coefficient:.2f}',
         transform=plt.gca().transAxes,
         fontsize=12,
         verticalalignment='top',
         bbox=dict(facecolor='white', alpha=0.5))
```

```python
#Make own column year
fandango_df['YEAR'] = fandango_df['FILM'].apply(lambda x: int(x.split()[-1][1:-1]) )
fandango_df['YEAR'].head()
fandango_df.head()
```

```python
#How many fandango movies are in the dataframe per year?
fandango_df.drop(columns=['RATING', 'STARS', 'VOTES']).groupby('YEAR').count()
```

```python
plt.figure()

count_plot = sns.countplot(data=fandango_df, x='YEAR', palette='dark')
```

```python
#What are the 10 movies with the highest number of votes
highest_votes_df = fandango_df.nlargest(10, 'VOTES')
highest_votes_df
```

```python
#How many movies have 0 votes?
votes_0 = len(fandango_df[fandango_df['VOTES'] == 0])
print(votes_0)
```

```python

fandango_df_0 = fandango_df[fandango_df['VOTES'] > 0].copy()
fandango_df_0
```

```python
#Create a distribution plot that shows the difference between the STARS and the TRUE rating on fandango
plt.figure(figsize=(12, 5))
sns.kdeplot(data=fandango_df_0, x='STARS', label='STARS', shade=True)
sns.kdeplot(data=fandango_df_0, x='RATING', label='RATING', shade=True)
plt.legend()

plt.show()
```

```python
#Create a column in the dataframe that is (STARS-RATING)
fandango_df_0['STARS DIFF'] = fandango_df_0.apply(lambda row: row['STARS'] - row['RATING'], axis=1)
fandango_df_0['STARS DIFF'] = fandango_df_0['STARS DIFF'].apply(lambda x: round(x, 1))
fandango_df_0.info()
```

```python
fandango_df_0['STARS DIFF'].unique()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=fandango_df_0, x='STARS DIFF', palette='dark')
```

```python
all_sites_df['RottenTomatoes-RottenTomatoes_User'] = all_sites_df.apply(lambda row: row['RottenTomatoes'] - row['RottenTomatoes_User'], axis=1)
all_sites_df.head()
```

```python
#Calculate the mean absoulte difference
all_sites_df['Abs Diff RottenTomatoes'] = all_sites_df['RottenTomatoes-RottenTomatoes_User'].apply(lambda x: abs(x))
all_sites_df.head()

```

```python
plt.figure(figsize=(12, 5))
sns.displot(data=all_sites_df, x='RottenTomatoes-RottenTomatoes_User', kde=True, aspect=2, bins=22, linewidth=1.5)
```

```python
plt.figure(figsize=(12, 5))
sns.displot(data=all_sites_df, x='Abs Diff RottenTomatoes', kde=True, aspect=2, bins=25)
plt.title('Abs Diff RottenTomatoes')
```

```python
#Display the movies that users love but critics hate
all_sites_df.head()
biggest_difference = all_sites_df['RottenTomatoes-RottenTomatoes_User'].nlargest(5)
print(biggest_difference)
biggest_difference_films = all_sites_df.loc[biggest_difference.index, 'FILM']
biggest_difference_films

```

```python
#Now do the movies that critics love but users hate
smallest_difference_films = all_sites_df.nsmallest(5, 'RottenTomatoes-RottenTomatoes_User')[['FILM', 'RottenTomatoes-RottenTomatoes_User']]
smallest_difference_films
```

```python
#Display metacritic user vs critic ratings
coeff = all_sites_df[['Metacritic', 'Metacritic_User']].corr()['Metacritic']['Metacritic_User']
coeff

plt.figure(figsize=(12, 5))
sns.scatterplot(data=all_sites_df, x='Metacritic', y='Metacritic_User', label=f'R={coeff: .2f}')
plt.legend(loc='upper left')
```

#Fandango vs Other Sites

```python
fandango_all_sites_df = pd.merge(all_sites_df, fandango_df, how='inner', on='FILM')
fandango_all_sites_df.info()
```

```python
fandango_all_sites_df.head()
```

```python
#Normalize columns so they are on the same rating scale as fandangos rating, (1-5)
norm_fandango_all_sites_df = fandango_all_sites_df.copy()

norm_fandango_all_sites_df['RottenTomatoes'] = fandango_all_sites_df['RottenTomatoes'].apply(lambda x: round(x/20, 1))
norm_fandango_all_sites_df['RottenTomatoes_User'] = fandango_all_sites_df['RottenTomatoes_User'].apply(lambda x: round(x/20, 1))
norm_fandango_all_sites_df['Metacritic'] = fandango_all_sites_df['Metacritic'].apply(lambda x: round(x/20, 1))
norm_fandango_all_sites_df['Metacritic_User'] = fandango_all_sites_df['Metacritic_User'].apply(lambda x: round(x/2, 1))
norm_fandango_all_sites_df['IMDB'] = fandango_all_sites_df['IMDB'].apply(lambda x: round(x/2, 1))

norm_fandango_all_sites_df.head()


```

```python
#How does the distribution of fandango's reviews compare to the other movie reviewers
plt.figure(figsize=(12, 5))

sns.kdeplot(data=norm_fandango_all_sites_df, x='Metacritic', shade=True, label='Metacritic')
sns.kdeplot(data=norm_fandango_all_sites_df, x='Metacritic_User', shade=True, label='Metacritic_User')
sns.kdeplot(data=norm_fandango_all_sites_df, x='IMDB', shade=True, label='IMDB')
sns.kdeplot(data=norm_fandango_all_sites_df, x='RottenTomatoes', shade=True, label='Rotten Tomatoes')
sns.kdeplot(data=norm_fandango_all_sites_df, x='RottenTomatoes_User', shade=True, label='Rotten Tomatoes User')

#Fandango plots
sns.kdeplot(data=norm_fandango_all_sites_df, x='STARS', shade=True, label='Fandango Stars', linewidth=2)
sns.kdeplot(data=norm_fandango_all_sites_df, x='RATING', shade=True, label='Fandango Rating', linewidth=2)


plt.legend()
```

```python
#Compare rotten tomatoes to Fandango ratings
plt.figure(figsize=(12, 5))

sns.kdeplot(data=norm_fandango_all_sites_df, x='RottenTomatoes', shade=True, label='Rotten Tomatoes')
sns.kdeplot(data=norm_fandango_all_sites_df, x='STARS', shade=True, label='Fandango Stars')

plt.legend(loc='upper left')
plt.xlim(0, 5)

plt.show()
```

```python
#Compare all the reviews with a histogram
plt.figure(figsize=(12, 5))
sns.histplot(norm_fandango_all_sites_df.drop(['IMDB_user_vote_count', 'RottenTomatoes-RottenTomatoes_User', 'YEAR', 'VOTES', 'Abs Diff RottenTomatoes', 'Metacritic_user_vote_count'], axis=1), bins=50)
```

```python
norm_fandango_all_sites_df.head()
```

```python
#Clustermap
cluster_df = norm_fandango_all_sites_df.drop(['RottenTomatoes-RottenTomatoes_User', 'Abs Diff RottenTomatoes', 'VOTES', 'YEAR', 'FILM', 'Metacritic_user_vote_count', 'IMDB_user_vote_count'], axis=1)

plt.figure()

sns.clustermap(data=cluster_df, col_cluster=False)
```

```python
#Find the 10 worst movies (based of rotten tomatoes critic ratings)
temp_df = norm_fandango_all_sites_df.drop(['RottenTomatoes-RottenTomatoes_User', 'Abs Diff RottenTomatoes', 'VOTES', 'YEAR', 'Metacritic_user_vote_count', 'IMDB_user_vote_count'], axis=1)
worst_movies = temp_df.nsmallest(10, 'RottenTomatoes')
worst_movies.set_index('FILM')
```

```python
#Compare the distributions across all sites for the worst movies

plt.figure(figsize=(12, 5))

sns.kdeplot(data=worst_movies, x='RottenTomatoes', shade=True, label='RottenTomatoes', palette='Set1', color='brown')
sns.kdeplot(data=worst_movies, x='RottenTomatoes_User', shade=True, label='RottenTomatoes_User', color='purple')
sns.kdeplot(data=worst_movies, x='Metacritic', shade=True, label='Metacritic')
sns.kdeplot(data=worst_movies, x='Metacritic_User', shade=True, label='Metacritic_User')
sns.kdeplot(data=worst_movies, x='IMDB', shade=True, label='IMDB')

sns.kdeplot(data=worst_movies, x='STARS', shade=True, label='STARS', linewidth=2, color='red')
sns.kdeplot(data=worst_movies, x='RATING', shade=True, label='RATING', linewidth=2, color='blue')

plt.legend()
plt.xlabel('Worst Movie Distributions')


```
